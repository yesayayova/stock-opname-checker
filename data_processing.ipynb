```python
df = pd.read_excel('data/Rekap Mutasi DK  (25.10.23).xlsx', sheet_name='REKAP MUTASI')

for i in range(7):
    df.drop(i, inplace=True)
df.reset_index(drop=True, inplace=True)

new_columns = df.iloc[0].values
new_columns[1] = 'PLU Number Barcode'
new_columns[2] = 'PLU Name Item Menu'
df.columns = new_columns 
df.drop(0, inplace=True)
df.reset_index(drop=True, inplace=True)

id_nan = []

for i in range(df.shape[0]):
    if type(df.iloc[i]['PLU Number Barcode']) != str:
        id_nan.append(i)
    
for id in id_nan:
    df.drop(id, inplace=True)

df.tail()
```
